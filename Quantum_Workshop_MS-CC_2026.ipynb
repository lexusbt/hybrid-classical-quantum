**Step 2: Executing the Circuit and plot the histogram of outcomes**

Now, execute the circuit using simulator AerSimulator (quantum simulator backend — a virtual quantum computer — using Qiskit Aer, which is Qiskit’s high-performance simulator framework). For this use .run() method.

 For execution,you may want to specify additional execution parameters such as the number of shots (repetitions of the experiment to gather statistics).


image.png

````python
#1 Import Necessary Modules
#Ensure the following imports are at the beginning of your notebook:

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator  # Updated simulator class
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# 2. Create quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply gates
qc.x(0)        # Apply X gate
qc.h(0)        # Apply Hadamard gate
qc.measure(0, 0) #measure

#3. Set Up the AerSimulator Backend Simulator
backend = AerSimulator()

#
# Transpilation adapts your circuit to the backend's requirements:

tqc = transpile(qc, backend)

# 4. Run the simulation
# Execute the transpiled circuit on the simulator, specifying the number of shots (repetitions):
job = backend.run(tqc, shots=1000)
result = job.result()

# 5. Retrieve and Display Measurement Results
# Obtain the counts of each measurement outcome:

counts = result.get_counts()
print(counts)

#6: Visualize the Results with a Histogram
#Plot the measurement outcomes:

plot_histogram(counts)
````

````python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator  # Use AerSimulator directly
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create a 2-qubit, 2-classical-bit circuit
qc = QuantumCircuit(2, 2)

# Apply gates
qc.h(0)        # Hadamard on q0
qc.cx(0, 1)    # CNOT from q0 to q1

# Measure both qubits
qc.measure(0, 0)
qc.measure(1, 1)

# Use the AerSimulator backend (modern approach)
backend = AerSimulator()

# Transpile for the simulator backend
tqc = transpile(qc, backend)

# Run simulation
job = backend.run(tqc, shots=1024)
result = job.result()

# Get and display counts
counts = result.get_counts()

# Plot the histogram
counts = result.get_counts()
print(counts)
plot_histogram(counts)

````
